**Project Overview**

This project analyzes UK-based online retail transaction data to understand
customer purchasing behavior, product performance, and revenue drivers.
The objective is to generate actionable business insights and build predictive
models that support revenue forecasting, demand planning, customer retention,
and product performance optimization.

**Business Objectives**

Improve revenue forecasting accuracy to support financial planning.

Identify high-performing and high-engagement products for inventory and promotion optimization.

Detect products and customers with high churn risk.

Enable data-driven decision-making using machine learning models.

**Regression Problem Statements**

1. Predict Next Month Product Sales

        Business Question:
        Can we predict the quantity of products sold in the next month using historical
        sales trends, pricing patterns, and seasonal effects?

        Target Variable:
        Monthly product sales quantity

        Business Value:
        Improves inventory planning and demand forecasting.

2. Forecast Revenue

       Business Question:
       Can future revenue be forecasted using historical sales quantities, pricing, and
       seasonality features?

       Target Variable:
       Monthly revenue
 
       Business Value:
       Supports revenue planning and business target setting.

3. Predict Future Demand

        Business Question:
        Can future demand levels be predicted to avoid overstocking or stock shortages?

        Target Variable:
        Future demand quantity

        Business Value:
        Reduces operational costs and improves supply chain efficiency

**Classification Problem Statements**

1. Classify Top-Selling vs Normal Products

        Business Question:
        Can products be classified as top-selling or normal based on historical sales
        volume, revenue contribution, and customer engagement?

        Target Variable:
        Product category (Top Seller / Normal)

        Business Value:
        Enables focused marketing and promotion strategies.

2. Predict High Churn vs Low Churn Products

        Business Question:
        Can products with declining demand be identified early using sales consistency
        and engagement patterns?

        Target Variable:
        Product churn category (High / Low)

        Business Value:
        Helps reduce revenue loss by identifying underperforming products early.



3. Predict High Engagement Products

       Business Question:
       Can we predict whether a product will have high customer engagement based on
       purchase frequency, active sales months, and customer reach?

       Target Variable:
       Product engagement level (High / Low)

       Business Value:
       Guides product prioritization and promotional strategy.
